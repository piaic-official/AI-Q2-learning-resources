```python
import numpy as np
import pandas as pd
import glob
import matplotlib.pyplot as plt
```

```python
files = glob.glob('./*.csv')
census = []
for file in files:
    c_df = pd.read_csv(file, index_col=None, header=0)
    census.append(c_df)
census_df = pd.concat(census, ignore_index=True)
display(census_df.head(10))
```

```python
census_df= census_df.drop(columns=['Unnamed: 0']) #deleting unnabed column
display(census_df.head())
```

```python
#splitting Gender column in two columns i.e. male and female
census_df[['Male','Female']] = census_df.GenderPop.str.split('_',expand=True)
census_df.drop(columns=['GenderPop'], inplace=True) #droping columng GdenderPop
```

```python
census_df.head()
```

```python
census_df['Income']= census_df.Income.str.strip('$') # removing '$' sign from income column
census_df= census_df.replace('%', '',regex=True) # removing % from the dataframe
census_df= census_df.replace('M', '',regex=True)
census_df= census_df.replace('F', '',regex=True)
census_df.head()
```

```python
census_df.dtypes
```

```python
census_df.loc[:,'Hispanic':'Income'] =round(census_df.loc[:,'Hispanic':'Income'].apply(pd.to_numeric),2)
```

```python
print(census_df.dtypes)
display(census_df.head())
```

```python
census_df['Male']= census_df['Male'].astype(int)
census_df
```

```python
census_df['Female']= census_df['TotalPop']-census_df['Male'] # filling empty values in female column by obtaing TotalPop - MalePop
census_df
```

```python
# drawing graph between average income and Female Pop
plt.scatter(census_df['Female'],census_df['Income'])
plt.xlabel('Income $',fontsize=16)
plt.ylabel('FemalePop', fontsize=16)
plt.show()
```

```python
#Checking Duplicate values 
census_df.duplicated().any()
```

```python
census_df.drop_duplicates(inplace =True)
```

```python
plt.scatter(census_df['Female'],census_df['Income'])
plt.xlabel('Income $',fontsize=16)
plt.ylabel('FemalePop', fontsize=16)
plt.show()
```
